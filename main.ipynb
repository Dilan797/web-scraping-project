```python

import pandas as pd
from urllib.parse import quote  # por si el nombre de liga lleva acentos o espacios
def obtener_estadisticas_equipos_fbref(
        pagina: str = "stats",
        id_liga: int = 8,
        nombre_liga: str = "Copa de la Liga Profesional",
        temporada: str | None = None,  
        guardar_csv: bool = False,
    ):
     # 1) Construir URL
    nombre_liga_mod = quote(nombre_liga.replace(" ", "-"))
    url = f"https://fbref.com/en/comps/{id_liga}/{pagina}/{nombre_liga_mod}-Stats"

    # 2) Leer todas las tablas de la página
    tablas = pd.read_html(url)
    if len(tablas) < 2:
        raise ValueError(f"Se esperaban ≥2 tablas, se encontraron {len(tablas)}")
    df_equipos, df_vs_equipos = tablas[0].copy(), tablas[1].copy()
    # 3) Limpiar encabezados si son MultiIndex
    for df in (df_equipos, df_vs_equipos):
        if isinstance(df.columns, pd.MultiIndex):
            df.columns = df.columns.droplevel(0)          # o get_level_values(-1)
        df.columns = df.columns.str.strip()
        df.insert(0, "Season", temporada or "current")  
    # 4) Guardar CSV opcional
    if guardar_csv:
        df_equipos.to_csv(f"estadisticas_{pagina}_equipos.csv", index=False)
        df_vs_equipos.to_csv(f"estadisticas_{pagina}_vs_equipos.csv", index=False)

    return df_equipos, df_vs_equipos
```

```python
df_eq, df_vs = obtener_estadisticas_equipos_fbref(
    pagina="possession",          # o "stats", "shooting", etc.
    id_liga=8,
    nombre_liga="Liga de Campeones",
    temporada ="2024-2025",
    guardar_csv=False
)
# 1) Ver las primeras 20 filas
pd.set_option("display.max_rows", 100)   # None para sin límite
display(df_eq.head(30))                  # muestra las 20 primeras
```

```python
import pandas as pd
from urllib.parse import quote

def obtener_posiciones_fbref(
        id_liga: int,
        temporada: str,                    # ej. "2023-2024"
        nombre_liga: str,
        guardar_csv: bool = False
    ):
    """
    Devuelve el DataFrame con la tabla de posiciones (clasificación) de una liga
    y temporada en FBref.

    Parámetros
    ----------
    id_liga     : número  FBref (Premier = 9, Liga MX = 31, etc.)
    temporada   : string "YYYY-YYYY"
    nombre_liga : tal cual aparece en la URL (sin 'Stats'), puede llevar espacios
    guardar_csv : si True, salva 'standings_<liga>_<temporada>.csv'
    """
    # normaliza nombre para la URL
    liga_url = quote(nombre_liga.replace(" ", "-"))
    url = f"https://fbref.com/en/comps/{id_liga}/{temporada}/{liga_url}-Stats"

    # 1️⃣  lee todas las tablas; busca la que contenga "Pts"
    tablas = pd.read_html(url, match="Pts", flavor="lxml")
    if not tablas:
        raise RuntimeError("No se detectó una tabla de standings en la página.")

    df_standings = tablas[0].copy()

    # 2️⃣  limpia encabezados multi-nivel, si los hubiera
    if isinstance(df_standings.columns, pd.MultiIndex):
        df_standings.columns = df_standings.columns.get_level_values(-1)

    df_standings.columns = df_standings.columns.str.strip()   # quita espacios
    df_standings.insert(0, "Season", temporada)

    # 3️⃣  guarda CSV opcional
    if guardar_csv:
        fname = f"standings_{liga_url}_{temporada}.csv"
        df_standings.to_csv(fname, index=False)
        print(f"📁  Clasificación guardada en {fname}")

    return df_standings
```

```python
df_tabla = obtener_posiciones_fbref(
    id_liga=8,
    temporada="2024-2025",
    nombre_liga="Champions League",
    guardar_csv=False
)

# mostrar las primeras 10 posiciones
pd.set_option("display.max_rows", 20)
df_tabla.head(20)
```
